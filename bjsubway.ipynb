```python
import requests 
from bs4 import BeautifulSoup
```

```python
import re
```

```python
def get_url(url, pattern):
    headers = {"User-Agent" : "User-Agent:Mozilla/5.0 (compatible; MSIE 9.0; Windows NT 6.1; Trident/5.0;"}
    r = requests.get(url, headers=headers).content.decode('utf8')
    
    soup = BeautifulSoup(r, "html.parser")
    l_can = soup.find_all('a')
#     print(l_can)
    
    sub_urls = []
    sub_names = []
    for s in l_can:
        print(s.string)
        if s.string and s.get('target') == '_blank':
            if re.findall(pattern, s.string):
                url = s.get('href')
                name = s.string
                if name not in sub_names:
                    sub_urls.append('https://baike.baidu.com'+ url)
                    sub_names.append(name)
#                 print(s.string, 'https://baike.baidu.com'+ url) 
    return zip(sub_urls, sub_names)
```

```python
url_bjsubwayline = 'https://baike.baidu.com/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485'
pattern_bjsubwayline ='^北京地铁.+线'
bjsubwaylines = get_url(url_bjsubwayline, pattern_bjsubwayline)
list(bjsubwaylines)
```

```python
l_stations = {}
for line_url, line_name in bjsubwaylines:
    stations_n_u = list(get_url(line_url, '.+站$'))
#     _, stations = list(zip(*stations_n_u))
    l_stations[line_name] = stations_n_u
```
